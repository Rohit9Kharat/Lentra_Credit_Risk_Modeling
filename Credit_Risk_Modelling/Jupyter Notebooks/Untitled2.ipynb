```python
def get_unique_values(values):
    unique = []
    for piece in values:
        if piece not in unique:
            unique.append(piece)
    return unique

id_list = list(df['ID'].values)
unq_id_list = get_unique_values(id_list)
```

```python
filename = 'b.csv'

for chunk in pd.read_csv(filename, chunksize=1000):
    #print('type: %s shape %s' % (type(chunk), chunk.shape))
    start_time = time.time()
    
    std_last1_mon = list()
    std_last3_mon_diff = list()
    std_last3_mon = list()
    std_last6_mon_diff = list()
    std_last6_mon = list()
    std_last1_yr_diff = list()
    std_last1_yr = list()
    std_last2_yr_diff = list()
    std_last2_yr = list()
    std_last3_yr_diff = list()
    std_last3_yr = list()
    matching = list()


    for x in range(0,chunk.shape[0]):
        if (chunk['mo_diff_ld_drc'][x]==-1):
            std_last1_mon.append('unknown')
        elif (chunk['mo_diff_ld_drc'][x]==1) and (chunk['payHistComp'][x][0]=='STD'):
            std_last1_mon.append(1)
        else:
            std_last1_mon.append(0)

        if (chunk['mo_diff_ld_drc'][x]==-1):
            std_last3_mon_diff.append('unknown')
        elif (chunk['mo_diff_ld_drc'][x]>1) and (chunk['mo_diff_ld_drc'][x]<4):
            std_last3_mon_diff.append(1)
        else:
            std_last3_mon_diff.append(0)

        if (chunk['mo_diff_ld_drc'][x]==-1):
            std_last6_mon_diff.append('unknown')
        elif (chunk['mo_diff_ld_drc'][x]>1) and (chunk['mo_diff_ld_drc'][x]<7):
            std_last6_mon_diff.append(1)
        else:
            std_last6_mon_diff.append(0)

        if (chunk['mo_diff_ld_drc'][x]==-1):
            std_last1_yr_diff.append('unknown')
        elif (chunk['mo_diff_ld_drc'][x]>1) and (chunk['mo_diff_ld_drc'][x]<13):
            std_last1_yr_diff.append(1)
        else:
            std_last1_yr_diff.append(0)

        if (chunk['mo_diff_ld_drc'][x]==-1):
            std_last2_yr_diff.append('unknown')
        elif (chunk['mo_diff_ld_drc'][x]>1) and (chunk['mo_diff_ld_drc'][x]<25):
            std_last2_yr_diff.append(1)
        else:
            std_last2_yr_diff.append(0)

        if (chunk['mo_diff_ld_drc'][x]==-1):
            std_last3_yr_diff.append('unknown')
        elif (chunk['mo_diff_ld_drc'][x]>1) and (chunk['mo_diff_ld_drc'][x]<37):
            std_last3_yr_diff.append(1)
        else:
            std_last3_yr_diff.append(0)

    chunk['std_last3_mon_diff'] = pd.Series(std_last3_mon_diff).values
    chunk['std_last6_mon_diff'] = pd.Series(std_last6_mon_diff).values
    chunk['std_last1_yr_diff'] = pd.Series(std_last1_yr_diff).values
    chunk['std_last2_yr_diff'] = pd.Series(std_last2_yr_diff).values
    chunk['std_last3_yr_diff'] = pd.Series(std_last3_yr_diff).values

    for x in range(0, chunk.shape[0]):
        if (chunk['std_last3_mon_diff'][x]=='unknown'):
            std_last3_mon.append('unknown')
        elif (chunk['std_last3_mon_diff'][x]==1):
            for i in range(0, 3-chunk['mo_diff_ld_drc'][x]):
                #print(i)
                matching.append(chunk['payHistComp'][x][i])
            if any('STD' in s for s in matching):
                std_last3_mon.append(1)
            else:
                std_last3_mon.append(0)
            #print(matching)    
            matching = list()
        else:
            std_last3_mon.append(0)

        if (chunk['std_last6_mon_diff'][x]=='unknown'):
            std_last6_mon.append('unknown')
        elif (chunk['std_last6_mon_diff'][x]==1):
            for i in range(0, 6-chunk['mo_diff_ld_drc'][x]):
                #print(i)
                matching.append(chunk['payHistComp'][x][i])
            if any('STD' in s for s in matching):
                std_last6_mon.append(1)
            else:
                std_last6_mon.append(0)
            #print(matching)    
            matching = list()
        else:
            std_last6_mon.append(0)

        if (chunk['std_last1_yr_diff'][x]=='unknown'):
            std_last1_yr.append('unknown')
        elif (chunk['std_last1_yr_diff'][x]==1) and (len(chunk['payHistComp'][x])>=12):
            for i in range(0, 12-chunk['mo_diff_ld_drc'][x]):
                #print(i)
                matching.append(chunk['payHistComp'][x][i])
            if any('STD' in s for s in matching):
                std_last1_yr.append(1)
            else:
                std_last1_yr.append(0)
            #print(matching)    
            matching = list()
        else:
            std_last1_yr.append(0)

        if (chunk['std_last2_yr_diff'][x]=='unknown'):
            std_last2_yr.append('unknown')
        elif (chunk['std_last2_yr_diff'][x]==1) and (len(chunk['payHistComp'][x])>=24):
            for i in range(0, 24-chunk['mo_diff_ld_drc'][x]):
                #print(i)
                matching.append(chunk['payHistComp'][x][i])
            if any('STD' in s for s in matching):
                std_last2_yr.append(1)
            else:
                std_last2_yr.append(0)
            #print(matching)    
            matching = list()
        else:
            std_last2_yr.append(0)

        if (chunk['std_last3_yr_diff'][x]=='unknown'):
            std_last3_yr.append('unknown')
        elif (chunk['std_last3_yr_diff'][x]==1) and (len(chunk['payHistComp'][x])==36):
            for i in range(0, 36-chunk['mo_diff_ld_drc'][x]):
                #print(i)
                matching.append(chunk['payHistComp'][x][i])
            if any('STD' in s for s in matching):
                std_last3_yr.append(1)
            else:
                std_last3_yr.append(0)
            #print(matching)    
            matching = list()
        else:
            std_last3_yr.append(0)

    chunk['STD1M_flag'] = pd.Series(std_last1_mon).values
    #print(chunk['STD1M_flag'].value_counts()) 

    chunk['STD3M_flag'] = pd.Series(std_last3_mon).values
    #print(chunk['STD3M_flag'].value_counts())

    chunk['STD6M_flag'] = pd.Series(std_last6_mon).values
    #print(chunk['STD6M_flag'].value_counts())

    chunk['STD1Y_flag'] = pd.Series(std_last1_yr).values
    #print(chunk['STD1Y_flag'].value_counts())

    chunk['STD2Y_flag'] = pd.Series(std_last2_yr).values
    #print(chunk['STD2Y_flag'].value_counts())

    chunk['STD3Y_flag'] = pd.Series(std_last3_yr).values
    #print(chunk['STD3Y_flag'].value_counts())

    print("%s seconds" % (time.time() - start_time))
```
