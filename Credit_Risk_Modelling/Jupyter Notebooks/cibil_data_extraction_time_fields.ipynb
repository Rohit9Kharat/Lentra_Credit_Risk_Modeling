```python
def search_dt_list(dateOpenedDisbursed_start,date_before_period, dateOpenedDisbursedlist):
    count=0
    for p in dateOpenedDisbursedlist:
        if dateOpenedDisbursed_start!=p and dateOpenedDisbursed_start > p >= date_before_period:
            count = count + 1
    return count
```

```python
for i in range(data.shape[0]):
    opened_last3M_CNT_P_list = []
    opened_last6M_CNT_P_list = []
    opened_last9M_CNT_P_list = []
    opened_last1Y_CNT_P_list = []
    opened_last2Y_CNT_P_list = []
    opened_last3Y_CNT_P_list = []
    
    for j in data['dateOpenedOrDisbursed'][i]:
        date_before_3M = j + three_mon_rel
        date_before_6M = j + six_mon_rel
        date_before_9M = j + nine_mon_rel
        date_before_1Y = j + one_year_rel
        date_before_2Y = j + two_year_rel
        date_before_3Y = j + three_year_rel
        
        
        
        opened_last3M_CNT_P_list.append(search_dt_list(j,date_before_3M, data['dateOpenedOrDisbursed'][i]))
        opened_last6M_CNT_P_list.append(search_dt_list(j,date_before_6M, data['dateOpenedOrDisbursed'][i]))
        opened_last9M_CNT_P_list.append(search_dt_list(j,date_before_9M, data['dateOpenedOrDisbursed'][i]))
        opened_last1Y_CNT_P_list.append(search_dt_list(j,date_before_1Y, data['dateOpenedOrDisbursed'][i]))
        opened_last2Y_CNT_P_list.append(search_dt_list(j,date_before_2Y, data['dateOpenedOrDisbursed'][i]))
        opened_last3Y_CNT_P_list.append(search_dt_list(j,date_before_3Y, data['dateOpenedOrDisbursed'][i]))
        
    data.iat[i,data.columns.get_loc('opened_last3M_CNT_P')] = opened_last3M_CNT_P_list
    data.iat[i,data.columns.get_loc('opened_last6M_CNT_P')] = opened_last6M_CNT_P_list
    data.iat[i,data.columns.get_loc('opened_last9M_CNT_P')] = opened_last9M_CNT_P_list
    data.iat[i,data.columns.get_loc('opened_last1Y_CNT_P')] = opened_last1Y_CNT_P_list
    data.iat[i,data.columns.get_loc('opened_last2Y_CNT_P')] = opened_last2Y_CNT_P_list
    data.iat[i,data.columns.get_loc('opened_last3Y_CNT_P')] = opened_last3Y_CNT_P_list
    

data.head()
```
